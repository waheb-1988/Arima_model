```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline

from math import sqrt
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Analysis imports
from pandas.plotting import lag_plot
from pylab import rcParams
from statsmodels.tsa.seasonal import seasonal_decompose
from pandas import DataFrame
from pandas import concat
```

```python
import pathlib
import os
```

```python

dir_path = pathlib.Path.cwd()
folder_path = dir_path / "data" / "air_passenger"
file_name = "airPassengers.csv"
df= pd.read_csv(os.path.join(folder_path, file_name))
print(df)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(22,6))
sns.lineplot(x=df.Month, y=df['#Passengers'])
plt.title('The Variation # of visit from 06/2017 until 09/2017')
plt.show()
```

```python
# i'm going to create a pivot table to plot the monthly temperatures through the years
df['day'] = df.index.day
df['month'] = df.index.month
pivot = pd.pivot_table(df, values='visit', index='month', columns='day', aggfunc='mean')
pivot.plot(figsize=(20,6))
plt.title('Monthly Visitors ')
plt.xlabel('days')
plt.ylabel('Visitors')
plt.xticks([x for x in range(1,13)])
plt.legend().remove()
plt.show()
```

```python
monthly_seasonality = pivot.mean(axis=1)
monthly_seasonality.plot(figsize=(20,6))
plt.title('Monthly Temperatures in Rio de Janeiro')
plt.xlabel('Months')
plt.ylabel('Temperature')
plt.xticks([x for x in range(1,13)])
plt.show()
```
